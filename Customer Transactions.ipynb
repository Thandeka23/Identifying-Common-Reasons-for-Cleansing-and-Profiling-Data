```python
import pandas as pd

# Load the dataset
file_path = r"C:\Users\giles\Downloads\customer_transactions.csv"
data = pd.read_csv(file_path)

# Display the first 10 rows
print("First 10 rows of the dataset:")
print(data.head(10))

# Summarize the structure of the dataset
summary = {
    'columns': data.columns.tolist(),
    'data_types': data.dtypes.to_dict(),
    'missing_values': data.isnull().sum().to_dict(),
    'missing_percentage': (data.isnull().sum() / len(data) * 100).to_dict()
}

print("\nSummary Report:")
print(f"Columns: {summary['columns']}")
print(f"Data Types: {summary['data_types']}")
print(f"Missing Values: {summary['missing_values']}")
print(f"Missing Percentage: {summary['missing_percentage']}")

# Identify columns with missing values and their percentage
missing_info = data.isnull().sum()
missing_percentage = (missing_info / len(data)) * 100
missing_columns = missing_info[missing_info > 0]

print("\nColumns with Missing Values:")
print(missing_columns)

# Fill missing values
for column in missing_columns.index:
    if data[column].dtype == 'object':  # Categorical column
        mode_value = data[column].mode()[0]
        data[column].fillna(mode_value, inplace=True)
        print(f"Filled missing values in '{column}' with mode: {mode_value}")
    else:  # Numerical column
        median_value = data[column].median()
        data[column].fillna(median_value, inplace=True)
        print(f"Filled missing values in '{column}' with median: {median_value}")

# Deliverable: Updated dataset with missing values handled
print("\nUpdated dataset after handling missing values:")
print(data.head(10))
```

```python
import pandas as pd

# Load the dataset
file_path = r"C:\Users\giles\Downloads\customer_transactions.csv"  # Adjust the path as necessary
data = pd.read_csv(file_path)

# Print the original column names
print("Original column names:")
print(data.columns)

# Check for duplicates
duplicate_count = data.duplicated().sum()
print(f"Number of duplicate rows found: {duplicate_count}")

# Remove duplicates
cleaned_data = data.drop_duplicates()

# Count how many were removed
removed_count = duplicate_count
print(f"Number of duplicate rows removed: {removed_count}")

# Standardize column names to uppercase
cleaned_data.columns = cleaned_data.columns.str.upper()

# Print the column names after standardization
print("Column names after standardization:")
print(cleaned_data.columns)

# Remove the '.1' from the 'TRANSACTION_AMOUNT.1' column name if it exists
cleaned_data.columns = cleaned_data.columns.str.replace('TRANSACTION_AMOUNT.1', 'TRANSACTION_AMOUNT', regex=False)

# Print the column names after renaming
print("Column names after renaming:")
print(cleaned_data.columns)

# Function to capitalize the first letter of each word
def capitalize_state_name(state):
    if isinstance(state, str):
        return state.title()  # Capitalizes the first letter of each word
    return state

# Apply the function to the 'STATE' column to ensure correct capitalization
cleaned_data['STATE'] = cleaned_data['STATE'].apply(capitalize_state_name)

# Check for duplicate 'TRANSACTION_AMOUNT' columns
transaction_amount_columns = [col for col in cleaned_data.columns if col == 'TRANSACTION_AMOUNT']

# If there are multiple 'TRANSACTION_AMOUNT' columns, drop the last one
if len(transaction_amount_columns) > 1:
    # Drop all but the first occurrence of 'TRANSACTION_AMOUNT'
    cleaned_data = cleaned_data.loc[:, cleaned_data.columns != 'TRANSACTION_AMOUNT']  # Remove all TRANSACTION_AMOUNT columns
    # Re-add the first occurrence of TRANSACTION_AMOUNT from the original data
    cleaned_data['TRANSACTION_AMOUNT'] = data['transaction_amount']  # Use the original column name

# Save the cleaned and standardized dataset to a new CSV file with a descriptive name
cleaned_file_path = r"C:\Users\giles\Downloads\customer_transaction_cleaned.csv"
cleaned_data.to_csv(cleaned_file_path, index=False)

print(f"Cleaned and standardized dataset saved to: {cleaned_file_path}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Step 1: Load the dataset
df = pd.read_csv(r'C:\Users\giles\Downloads\customer_transactions_cleaned_standardized 1.csv')

# Step 2: Visualize outliers using boxplots
# Create boxplots for numerical columns
numerical_columns = ['CUSTOMER_AGE', 'TRANSACTION_AMOUNT.1']  # Example numerical columns
for col in numerical_columns:
    plt.figure(figsize=(10, 5))
    sns.boxplot(x=df[col])
    plt.title(f'Boxplot for {col}')
    plt.show()

# Step 3: Handle outliers
# Define a function to replace outliers with the median
def replace_outliers_with_median(column):
    Q1 = column.quantile(0.25)
    Q3 = column.quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    # Replace outliers with the median
    median = column.median()
    column = column.where((column >= lower_bound) & (column <= upper_bound), median)
    return column

# Apply the function to numerical columns
for col in numerical_columns:
    df[col] = replace_outliers_with_median(df[col])

# Step 4: Verify changes
print("\nDataset after handling outliers:")
print(df.describe())

# Step 5: Save the cleaned dataset
df.to_csv(r'C:\Users\giles\Downloads\customer_transactions_cleaned_no_outliers.csv', index=False)
```
